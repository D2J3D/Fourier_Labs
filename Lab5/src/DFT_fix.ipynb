```python
import numpy as np
from numpy import sin, cos, exp, pi, sqrt
import matplotlib.pyplot as plt
```

```python
# Define square wave
def sq_wave(t):
    if np.abs(t) <= 0.5:
        return 1
    return 0
```

```python
# Define timeline and our function's timeline 
t0 = -10 #  defining timline = [t0, -t0]
N = 1000 # amount of the points on a timeline
timeline = np.linspace(t0, -t0, N)
sqWave_timeline = [sq_wave(t) for t in timeline]
```

```python
# Plot square wave and it's DFT 
fig, ax = plt.subplots()
ax.plot(timeline, sqWave_timeline)
ax.grid(True)
plt.title("Square wave")
None
```
